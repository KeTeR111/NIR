```python
from pathlib import Path
import pandas as pd

folder = Path("/Users/andrey/Desktop/NIR/Datasets")

for file_path in folder.rglob('*'):  # rglob для рекурсивного поиска
    if file_path.is_file():
        print(f"Обрабатываю: {file_path}")
        
        try:
            df = pd.read_csv(file_path, sep='\t', header=None, decimal=',')
            df.columns = ['X', 'Y']
            df.to_csv(file_path, sep='\t', index=False)
        except Exception as e:
            print(f"Ошибка: {e}")
```
